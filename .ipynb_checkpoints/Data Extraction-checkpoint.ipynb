```python
import csv
import requests
from bs4 import BeautifulSoup

# Function to extract article title and text from a URL
def extract_information(URL):
    try:
        # Send a GET request to the URL
        response = requests.get(URL)
        
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.content, "html.parser")
            
            # Extract article title
            title = soup.title.text.strip()
            
            # Find the main article content
            article_content = soup.find('div',class_ = 'td-post-content')
            
            # Extract article text
            if article_content:
                article_text = article_content.get_text(separator="\n")
            else:
                article_text = "No article content found"
            
            return {"title": title, "text": article_text}
        elif response.status_code == 404:
            pass
        else:
            print(f"Failed to fetch URL: {URL}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error occurred while processing URL: {URL}. Error: {e}")
        return None

# Open the CSV file to read URLs
with open("input.csv", "r") as csvfile:
    # Create a CSV reader object
    csvreader = csv.reader(csvfile)
    
    # Skip the header row
    next(csvreader)
    
    # Store extracted data
    extracted_data = []
    
    # Iterate over each row in the CSV file
    for row in csvreader:
        URL_ID, URL = row
        
        # Extract information from the URL
        info = extract_information(URL)
        
        # If information is extracted successfully, append it to extracted_data
        if info:
            extracted_data.append({"URL_ID": URL_ID, "URL": URL, "title": info["title"], "text": info["text"]})

# Write the extracted data into a CSV file
#with open("Output Data Structure.csv", "w", newline="", encoding="utf-8") as csvfile:
    # Define the CSV writer object
#    csvwriter = csv.DictWriter(csvfile, fieldnames=["URL_ID", "URL", "title", "text"])
    
    # Write the header row
#    csvwriter.writeheader()
    
    # Write the extracted data
#    csvwriter.writerows(extracted_data)
```
